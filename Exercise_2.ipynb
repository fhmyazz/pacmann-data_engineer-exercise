### **1**

Buatlah DataFrame dari file yang sudah diekstrak tersebut

```python
# Read data
car_data = pd.read_csv('car_price.csv')

# Menampilkan 5 data teratas
car_data.head()
```

### **2**
---

Tampilkan jumlah baris dan kolom dari data tersebut

```python
# Jumlah baris
car_data_row = car_data.shape[0]

# Jumlah kolom
car_data_col = car_data.shape[1]

# Print jumlah baris dan kolom
print(f'Jumlah baris: {car_data_row}')
print(f'Jumlah kolom: {car_data_col}')
```

### **3**
---

Coba hapus kolom **Unnamed: 0**

```python
# Menghapus kolom 'Unnamed: 0'
car_data = car_data.drop('Unnamed: 0', axis=1)

# Tampilkan 5 data teratas
car_data.head()
```

### **4**
---

- Tampilkan tipe-tipe data dati setiap kolom

```python
# Menampilkan tipe-tipe data dari setiap kolom
car_data.dtypes
```

### **5**
---

- Pada kolom `car_prices_in_rupee`, rubahlah tipe datanya menjadi **float**
- Oleh karena itu, **hapus** terlebih dahulu **string "Lakh" dsb** pada kolom tersebut.

- Hapus string "Lakh" dsb

```python
# Menghapus tanda string "Lakh" dsb dari kolom car_prices_in_rupee
replace_rupee_price = {
    ' Lakh': '',
    ' Crore': '',
    ',': '',
    ' ': ''
}
car_data['car_prices_in_rupee'] = car_data['car_prices_in_rupee'].replace(replace_rupee_price, regex=True)

# Tampilkan 5 data teratas
car_data.head()
```

- Rubah tipe datanya menjadi float

```python
# Melakukan casting ke float
car_data['car_prices_in_rupee'] = car_data['car_prices_in_rupee'].astype('float')

# Tampilkan tipe data dari setiap kolom
car_data.dtypes
```

### **6**
---

- Pada kolom `kms_driven`, rubahlah tipe datanya menjadi **integer**
- Oleh karena itu, **hapus** terlebih dahulu **string "kms" dsb** pada kolom tersebut.

- Hapus string "kms" dsb

```python
# Menghapus tanda string "kms" dsb dari kolom kms_driven
replace_kms_drive = {
    ' kms': '',
    ',': ''
}
car_data['kms_driven'] = car_data['kms_driven'].replace(replace_kms_drive, regex=True)

# Tampilkan 5 data teratas
car_data.head()
```

- Rubah tipe datanya menjadi integer

```python
# Melakukan casting ke integer
car_data['kms_driven'] = car_data['kms_driven'].astype('int64')

# Tampilkan tipe data dari setiap kolom
car_data.dtypes
```
