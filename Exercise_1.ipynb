## **Case 1: Transactional Data**
---

Diberikan sebuah tabel transactional data

<center>

| **User** | **Frequency** | **Total Spend (Rp)** |
|----------|---------------|----------------------|
|   Cahyo  |       5       |       3.500.000      |
|   Budi   |       1       |        125.000       |
|   Angga  |       4       |       1.250.000      |
|  Risman  |       8       |       5.550.000      |
|    Dio   |       3       |       3.500.000      |
|   Somad  |       10      |       4.250.000      |
|  Bambang |      15       |       7.500.000      |

</center>

### **1**
---

Buatlah DataFrame berdasarkan table tersebut

```python
# code in here
datas = {
    'User': ['Cahyo', 'Budi','Angga','Risman','Dio','Somad','Bambang'],
    'Frequency': [5, 1, 4, 8, 3, 10, 15],
    'Total Spend (Rp)': [3500000, 125000, 1250000, 5550000, 3500000, 4250000, 7500000]
}
df_datas = pd.DataFrame(datas)
df_datas
```

### **2**
---

Lakukan process select data yang memiliki kolom `user` dan `total_spend`

```python
# Code in here
df_datas[['User', 'Total Spend (Rp)']]
```

### **3**
---

Coba lakukan filter untuk mengambil data yang memiliki `frequency` lebih dari `3`

```python
# Code in here
df_datas[df_datas['Frequency'] > 3]
```

### **4**
---

- Lakukan filter data yang dimana:
    - nilai `frequency` harus lebih dari sama dengan `4`
    - nilai `total_spend` lebih dari `5_000_000`

```python
# Code in here
df_datas = df_datas[(df_datas['Frequency'] >= 4) & (df_datas['Total Spend (Rp)'] > 5000000)]
df_datas
```

### **5**
---

Simpanlah data terbaru tersebut dengan filename `new_transactional_data.csv`

```python
# Code in here
df_datas.to_csv("new_transactional_data.csv")
```

Mengecek apakah data yang terbaru sudah tersimpan atau belum

```python
pd.read_csv("new_transactional_data.csv")
```

## **Study Case 2: Football API Data**
---

### **1**
---

- Diberikan sebuah link API `https://api.football-data.org/v4/competitions/` yang isinya adalah data tentang Football
- Coba cek status code dari API tersebut dengan menggunakan library `requests`

```python
# Code in here
response = requests.get('https://api.football-data.org/v4/competitions/')
response
```

### **2**
---

Ubah API response tersebut ke dalam bentuk JSON, setelah itu print hasil nya

```python
# Code in here
json_datas = response.json()
json_datas
```

### **3**
---

- Ubah data JSON tersebut, menjadi bentuk DataFrame
- Pastikan memiliki output yang sama seperti di bawah ini

```python
# Code in here
football_datas = pd.DataFrame(json_datas['competitions'])
football_datas
```

### **4**
---

- Select data yang memiliki kolom - kolom berikut:
    - `id`
    - `name`
    - `code`
    - `type`
    - `plan`
    - `lastUpdated`

```python
# Code in here
football_datas = football_datas[['id','name','code','type','plan','lastUpdated']]
football_datas
```

### **5**
---

Ubahlah kolom `lastUpdated` menjadi `last_updated`

```python
# Code in here
rename_football_datas = {
    'lastUpdated':'last_updated'
}
football_datas = football_datas.rename(columns = rename_football_datas)
football_datas
```

### **6**
---

- Lakukanlah Filter Data:
    - kolom `type` memiliki value `CUP`
    - kolom `plan` memiliki value `TIER_TWO` atau `TIER_FOUR`

```python
# Code in here
football_datas[(football_datas['type'] == 'CUP') & ((football_datas['plan'] == 'TIER_TWO') | (football_datas['plan'] == 'TIER_FOUR'))]
```

### **7**
---

Simpanlah data terbaru tersebut dengan filename `new_football_data.csv`

```python
# Code in here
football_datas.to_csv('new_football_data.csv')
```

Cek apakah data sudah berhasil tersimpan atau belum

```python
pd.read_csv("new_football_data.csv")
```
